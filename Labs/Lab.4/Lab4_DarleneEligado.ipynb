1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self,max_value):
        self.max_value = max_value
        self.value = 0 
        
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("ERROR!! Counter has hit the maximim value!")
            
    def reset(self):
        self.value = 0 
        
    def count_value(self):
        return self.value    
```

```python
class rectangle(shapes): 
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def rect_length(self):
        return self.__length

    def rect_width(self):
        return self.__width

    def coor_x(self):
        return self.__x

    def coor_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def parameter(self, num_pairs = 16):
        pairs = []
        perimeter = self.perimeter()
        step = perimeter / num_pairs
        for i in range(num_pairs):
            angle = (i / num_pairs) * 2 * math.pi
            x = self.coor_x() + math.cos(angle) * (self.rect_length() if i < num_pairs / 2 else -self.rect_length())
            y = self.coor_y() + math.sin(angle) * self.rect_width()
            pairs.append((x, y))
        return pairs
    
    def inside(self, x, y): # check if the x & y corrs are in the rec's boundaries
        return (self.coor_x() <= x <= self.coor_x() + self.rect_length()) and (self.coor_y() <= y <= self.coor_y() + self.rect_width())
    
    def overlaps(self, other):
        if isinstance(other, rectangle):
            x_dist = self.coor_x() - other.coor_x()
            y_dist = self.coor_y() - other.coor_y()
            x_overlap = self.__length / 2 + other.__length / 2 >= x_dist and -self.__length / 2 <= x_dist
            y_overlap = self.__width / 2 + other.__width / 2 >= y_dist and -self.__width / 2 <= y_dist
            return x_overlap and y_overlap
        else:
            return False
```

```python
class shapes: 
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
        
    def area(self):
        raise NotImplementedError
    
    def perimeter(self):
        raise NotImplementedError
        
    def parameter(self, num_pairs = 16):
        raise NotImplementedError
        
    def inside(self, x, y):
        raise NotImplementedError
        
    def overlaps(self, other):
        raise NotImplementedError
####
class rectangle(shapes): 
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def rect_length(self):
        return self.__length

    def rect_width(self):
        return self.__width

    def coor_x(self):
        return self.__x

    def coor_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def parameter(self, num_pairs = 16):
        pairs = []
        perimeter = self.perimeter()
        step = perimeter / num_pairs
        for i in range(num_pairs):
            angle = (i / num_pairs) * 2 * math.pi
            x = self.coor_x() + math.cos(angle) * (self.rect_length() if i < num_pairs / 2 else -self.rect_length())
            y = self.coor_y() + math.sin(angle) * self.rect_width()
            pairs.append((x, y))
        return pairs
    
    def inside(self, x, y): # check if the x & y corrs are in the rec's boundaries
        return (self.coor_x() <= x <= self.coor_x() + self.rect_length()) and (self.coor_y() <= y <= self.coor_y() + self.rect_width())
    
    def overlaps(self, other):
        if isinstance(other, rectangle):
            x_dist = self.coor_x() - other.coor_x()
            y_dist = self.coor_y() - other.coor_y()
            x_overlap = self.__length / 2 + other.__length / 2 >= x_dist and -self.__length / 2 <= x_dist
            y_overlap = self.__width / 2 + other.__width / 2 >= y_dist and -self.__width / 2 <= y_dist
            return x_overlap and y_overlap
        else:
            return False
#### 
class circle(shapes):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y 
        
    def cir_radius(self):
        return self.__radius
    
    def coor_x(self):
        return self.__x

    def coor_y(self):
        return self.__y
    
    def area(self):
        return math.pi*self.__radius**2 
    
    def perimeter(self):
        return 2*math.pi*self.__radius
    
    def parameter(self, num_pairs = 16):
        pairs = []
        for i in range(num_pairs):
            angle = (i / num_pairs) * 2 * math.pi
            x = self.coor_x() + self.cir_radius() * math.cos(angle)
            y = self.coor_y() + self.cir_radius() * math.sin(angle)
            pairs.append((x, y))
        return pairs
    
    def inside(self, x, y):
        distance2 = (x - self.coor_x()) ** 2 + (y - self.coor_y()) ** 2 # distance btwn the given point & the center of the circle 
        return distance2 <= self.__radius ** 2 # test if distance is <= the square of the radius
    

    def overlaps(self, other):
        if isinstance(other, circle):
            dist2 = (self.coor_x() - other.coor_x())**2 + (self.coor_y() - other.coor_y())**2
            radius_sum2 = (self.__radius + other.__radius)**2
            return dist2 <= radius_sum2
        else:
            return False
####
class triangle(shapes):
    def __init__(self, a, b, c, x, y): #a-c are sides of triangle
            self.__a = a
            self.__b = b
            self.__c = c
            self.__x = x
            self.__y = y
            
    def a(self):
        return self.__a
    
    def b(self):
        return self.__b
    
    def c(self):
        return self.__c
    
    def coor_x(self):
        return self.__x

    def coor_y(self):
        return self.__y
    
    def area(self): # using heron's formula, A = sqrt(S(s-a)(s-b)(s-c))
        s = (self.a() + self.b() + self.c()) / 2
        return math.sqrt(s * (s - self.a()) * (s - self.b()) * (s - self.c()))
    
    def perimeter(self): # a + b + c
        return self.a() + self.b() + self.c()
    
    def parameter(self, num_pairs = 16): # returns vertices of the tri
        return [(self.coor_x(), self.coor_y()),
                (self.coor_x() + self.a(), self.coor_y()),
                (self.coor_x(), self.coor_y() + self.b())]
    
    def inside(self, x, y):
        area = self.area() # calculate the signed areas of the sub-tris by the point & each pair of vertices
        area1 = 0.5 * ((x - self.coor_x()) * (self.coor_y() + self.b()) - (self.coor_x() + self.a() - x) * (self.coor_y() - y))
        area2 = 0.5 * ((self.coor_x() + self.a() - x) * (self.coor_y() - y) - (self.coor_x() - x) * (self.coor_y() - y))
        area3 = 0.5 * ((self.coor_x() - x) * (self.coor_y() - y) - (x - self.coor_x()) * (self.coor_y() + self.b()))
        
        return math.isclose(area, area1 + area2 + area3) # test if sum of areas is == to the area of tri

    def overlaps(self, other):
        if isinstance(other, triangle): # checks if any vertex of one tri lies inside another tri
            for point in self.parameter():
                if other.inside(point[0], point[1]):
                    return True
            for point in other.parameter():
                if self.inside(point[0], point[1]):
                    return True
            return False
        else:
            return False
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        
    class CompoundShape(shapes):
        def __init__(self, shapes):
            self.shapes = shapes

        def overlaps(self, other):
            for shape in self.shapes:
                if shape.overlaps(other):
                    return True
            return False
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint(self):
        for shape in self.shapes:
            if isinstance(shape, rectangle):
                for x in range(shape.coor_x(), shape.coor_x() + shape.rect_length()):
                    for y in range(shape.coor_y(), shape.coor_y() + shape.rect_width()):
                        if shape.inside(x, y):
                            self.canvas.set_pixel(y, x)
            elif isinstance(shape, circle):
                for x in range(shape.coor_x() - shape.cir_radius(), shape.coor_x() + shape.cir_radius()):
                    for y in range(shape.coor_y() - shape.cir_radius(), shape.coor_y() + shape.cir_radius()):
                        if shape.inside(x, y):
                            self.canvas.set_pixel(y, x)
            elif isinstance(shape, triangle):
                for x in range(int(shape.coor_x()), int(shape.coor_x() + shape.a())):
                    for y in range(int(shape.coor_y()), int(shape.coor_y() + shape.b())):
                        if shape.inside(x, y):
                            self.canvas.set_pixel(y, x)

    def display(self):
        self.canvas.display()
```

```python
draw = RasterDrawing(20, 20)

# Adding shapes to the drawing
draw.add_shape(rectangle(5, 10, 2, 2))
draw.add_shape(circle(5, 15, 15))
draw.add_shape(triangle(5, 5, 5, 10, 5))


draw.paint()
draw.display()
```
